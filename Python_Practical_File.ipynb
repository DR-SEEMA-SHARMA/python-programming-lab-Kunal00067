```python
# Program 3:
def remove_duplicates(lst):
    result = []
    for item in lst:
        if item not in result:
            result.append(item)
    return result

# User input
a = list(map(int, input("Enter numbers separated by spaces: ").split()))
print("List without duplicates:", remove_duplicates(a))
```

```python
# Program 4:
import random
import string

def generate_password(length=12):
    characters = string.ascii_letters + string.digits + string.punctuation
    password = ''.join(random.choice(characters) for _ in range(length))
    return password

def main():
    while True:
        length = int(input("Enter password length: "))
        print("Generated Password:", generate_password(length))
        again = input("Do you want to generate another password? (yes/no): ").lower()
        if again != "yes":
            print("Goodbye!")
            break

if __name__ == "__main__":
    main()
```

```python
# Program 11:
def largest(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c

# User input
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))
c = float(input("Enter third number: "))

print("The largest number is:", largest(a, b, c))
```

```python
# Program 13:
# Base class
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print(f"ðŸ‘¤ Name: {self.name}, Age: {self.age}")

# Derived class
class Employee(Person):
    def __init__(self, name, age, salary):
        super().__init__(name, age)
        self.salary = salary

    def display(self):
        super().display()
        print(f"ðŸ’° Salary: ${self.salary:.2f}")

    # Operator overloading for +
    def __add__(self, other):
        return self.salary + other.salary


# User input for first employee
print("Enter details for the first employee:")
name1 = input("Name: ")
age1 = int(input("Age: "))
salary1 = float(input("Salary: "))

# User input for second employee
print("\nEnter details for the second employee:")
name2 = input("Name: ")
age2 = int(input("Age: "))
salary2 = float(input("Salary: "))

# Creating employee objects
emp1 = Employee(name1, age1, salary1)
emp2 = Employee(name2, age2, salary2)

# Display details
print("\nðŸ“‹ Employee 1 Information:")
emp1.display()

print("\nðŸ“‹ Employee 2 Information:")
emp2.display()

# Operator overloading demonstration
print(f"\nðŸ’¼ Combined Salary of Both Employees: ${emp1 + emp2:.2f}")
```

```python
# Program 14:
try:
    num1 = int(input("Enter first number: "))
    num2 = int(input("Enter second number: "))
    result = num1 / num2
    print("Result:", result)

except ValueError:
    print("Error: Please enter valid integers only.")

except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")

except Exception as e:
    print("An unexpected error occurred:", e)

else:
    print("Division successful!")

finally:
    print("Program execution completed.")
```
